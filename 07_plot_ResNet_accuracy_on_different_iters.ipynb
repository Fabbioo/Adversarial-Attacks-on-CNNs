# Load and display performances on different iters

```python
fgsm: list[float] = np.loadtxt("res/fgsm_accuracies.txt", delimiter = ' ').tolist()

plt.figure()
plt.plot(epsilons, fgsm, marker = 'o', color = 'r')
plt.suptitle("Performance del modello ResNet-50 contro l\'attacco FGSM al variarie di $\epsilon$")
plt.xlabel("Epsilon")
plt.ylabel("Accuracy")
plt.xticks(np.arange(0, 1.1, step = 0.1))
plt.yticks(np.arange(0, 1.1, step = 0.1))
plt.grid()

del fgsm
```

```python
ifgsm_5_iters: list[float] = np.loadtxt("res/ifgsm_accuracies_5iters.txt", delimiter = ' ').tolist()
ifgsm_10_iters: list[float] = np.loadtxt("res/ifgsm_accuracies_10iters.txt", delimiter = ' ').tolist()
ifgsm_25_iters: list[float] = np.loadtxt("res/ifgsm_accuracies_25iters.txt", delimiter = ' ').tolist()
ifgsm_50_iters: list[float] = np.loadtxt("res/ifgsm_accuracies_50iters.txt", delimiter = ' ').tolist()
ifgsm_75_iters: list[float] = np.loadtxt("res/ifgsm_accuracies_75iters.txt", delimiter = ' ').tolist()
ifgsm_100_iters: list[float] = np.loadtxt("res/ifgsm_accuracies_100iters.txt", delimiter = ' ').tolist()

plt.figure()
plt.plot(epsilons, ifgsm_5_iters, label = "iters = 5", marker = 'o', color = 'y')
plt.plot(epsilons, ifgsm_10_iters, label = "iters = 10", marker = 'o', color = 'g')
plt.plot(epsilons, ifgsm_25_iters, label = "iters = 25", marker = 'o', color = 'b')
plt.plot(epsilons, ifgsm_50_iters, label = "iters = 50", marker = 'o', color = 'm')
plt.plot(epsilons, ifgsm_75_iters, label = "iters = 75", marker = 'o', color = 'c')
plt.plot(epsilons, ifgsm_100_iters, label = "iters = 100", marker = 'o', color = 'r')
plt.legend(loc = "lower left")
plt.suptitle("Performance del modello ResNet-50 contro l\'attacco I-FGSM al variarie di $\epsilon$ e sfruttando differenti iters")
plt.xlabel("Epsilon")
plt.ylabel("Accuracy")
plt.xticks(np.arange(0, 1.1, step = 0.1))
plt.yticks(np.arange(0, 1.1, step = 0.1))
plt.grid()

del ifgsm_5_iters, ifgsm_10_iters, ifgsm_25_iters, ifgsm_50_iters, ifgsm_75_iters, ifgsm_100_iters
```
